```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import spearmanr, pearsonr
from datetime import datetime
```

```python
math_grades = [17.5 , 16, 17, 15, 15, 16, 18, 20, 13.5, 17, 16, 18, 13.5 , 20, 15.0, 16.5, 16.5, 15, 17, 18]
physics_grades = [20, 81, 80, 70, 85, 75, 87, 97, 35, 90, 83, 88, 69, 100, 81, 10, 79, 78, 84, 91]
```

```python
def calculate_statistics(data):
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  
    median = np.median(data)
    q1 = np.quantile(data, 0.25)
    q3 = np.quantile(data, 0.75)
    return mean, std_dev, median, q1, q3
```

```python

math_mean, math_std_dev, math_median, math_q1, math_q3 = calculate_statistics(math_grades)
physics_mean, physics_std_dev, physics_median, physics_q1, physics_q3 = calculate_statistics(physics_grades)

print("Math Grades:")
print(f"Mean: {math_mean}")
print(f"Standard Deviation: {math_std_dev}")
print(f"Median: {math_median}")
print(f"First Quantile (Q1): {math_q1}")
print(f"Third Quantile (Q3): {math_q3}")

print("\nPhysics Grades:")
print(f"Mean: {physics_mean}")
print(f"Standard Deviation: {physics_std_dev}")
print(f"Median: {physics_median}")
print(f"First Quantile (Q1): {physics_q1}")
print(f"Third Quantile (Q3): {physics_q3}")


fig, axes = plt.subplots(1, 2, figsize=(8, 3))


axes[0].hist(math_grades, bins=6, color='skyblue', edgecolor='black', alpha=0.7)
axes[0].axvline(math_mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {math_mean:.2f}')
axes[0].axvline(math_median, color='green', linestyle='dashed', linewidth=2, label=f'Median: {math_median:.2f}')
axes[0].set_title('Math Grades Histogram')
axes[0].set_xlabel('Grades')
axes[0].set_ylabel('Frequency')
axes[0].legend()


axes[1].hist(physics_grades, bins=6, color='lightcoral', edgecolor='black', alpha=0.7)
axes[1].axvline(physics_mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {physics_mean:.2f}')
axes[1].axvline(physics_median, color='green', linestyle='dashed', linewidth=2, label=f'Median: {physics_median:.2f}')
axes[1].set_title('Physics Grades Histogram')
axes[1].set_xlabel('Grades')
axes[1].set_ylabel('Frequency')
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def normalize(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    return [(x - mean) / std_dev for x in data]


math_grades = [17.5, 16, 17, 15, 15, 16, 18, 20, 13.5, 17, 16, 18, 13.5, 20, 15.0, 16.5, 16.5, 15, 17, 18]
physics_grades = [20, 81, 80, 70, 85, 75, 87, 97, 35, 90, 83, 88, 69, 100, 81, 10, 79, 78, 84, 91]


normalized_math = normalize(math_grades)
normalized_physics = normalize(physics_grades)


plt.figure(figsize=(10,5))
plt.plot(normalized_math, label='Normalized Math Grades', marker='o')
plt.plot(normalized_physics, label='Normalized Physics Grades', marker='s')
plt.axhline(y=0, color='gray', linestyle='--', linewidth=1)
plt.xlabel('Student Index')
plt.ylabel('Normalized Score')
plt.title('Normalized Math and Physics Grades')
plt.legend()
plt.show()
```

```python
def plot_qq(data, title):
    plt.figure(figsize=(4, 4))
    stats.probplot(data, dist="norm", plot=plt)
    plt.title(f'QQ Plot for {title}')
    plt.show()


plot_qq(math_grades, 'Math Grades')


plot_qq(physics_grades, 'Physics Grades')
```

```python

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

vectorizer = TfidfVectorizer()
tfidf_matrix_title = vectorizer.fit_transform(cleaned_df['title'])
tfidf_matrix_discip = vectorizer.fit_transform(cleaned_df['description'])
cosine_sim_title= cosine_similarity(tfidf_matrix_title)
cosine_sim_discip= cosine_similarity(tfidf_matrix_discip)

counter = 0

duplicated_rows = set()

for i in range(len(cleaned_df)):
    for j in range(i + 1, len(cleaned_df)):  
        if (
            cleaned_df.loc[i, 'propertysize'] == cleaned_df.loc[j, 'propertysize'] and
            cleaned_df.loc[i, 'totalprice'] == cleaned_df.loc[j, 'totalprice'] and
            cleaned_df.loc[i, 'roomcount'] == cleaned_df.loc[j, 'roomcount'] and
            cleaned_df.loc[i, 'buildyear'] == cleaned_df.loc[j, 'buildyear'] and
            cosine_sim_title[i, j] > 0.9 and 
            cosine_sim_discip[i, j] > 0.9 
        ):
            duplicated_rows.add(j)  
            print(f"Rows {i} and {j} are similar")


print(f"Befor removing duplicated advertisments :{cleaned_df.shape}")
cleaned_df = cleaned_df.drop(index=list(duplicated_rows)).reset_index(drop=True).copy()


print("New shape of cleaned_df:", cleaned_df.shape)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


def remove_outliers(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df


outlier_columns = ["propertysize", "totalprice_int"]


cleaned_df_filtered = remove_outliers(cleaned_df, outlier_columns)


plt.figure(figsize=(15, 10))
for i, column in enumerate(float_columns, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=cleaned_df_filtered[column], color='orange')
    plt.title(f'Boxplot of {column}')
    plt.ylabel(column)

plt.tight_layout()
plt.show()
```
